```python
import numpy as np
import pandas as pd
import glob
file_1 , file_2 = glob.glob('*csv*')
bank_df = pd.read_csv(file_2)
```

```python
bank_key = np.array(bank_df.keys())
bank_key
```

```python
X = bank_df.drop([bank_key[0],bank_key[4],bank_key[9]],axis=1)
X.head()
# X.shape
y = bank_df[bank_key[9]]
np.bincount(y)
```

```python
# one-hot encoding 과정 (문자형 숫자(범주)>> 범주형 변수>> 가변수로)
X['Education'] = X['Education'].replace([1,2,3],['Under','Grad','Prof'])
# categori로 나누려는대 숫자 형태로 되어있으면 분석시 오류가 생긴다 ex(혈액형,학년,반)
# 알파벳 기준으로 가장 낮은 값인 G로 시작하는 Grad가 기준값으로 사용된다
X.head()
```

```python
# 가변수 변환
X=pd.get_dummies(X[X.columns],columns=['Education'],drop_first=True)
# drop_first >> 기준값을 살릴 것인지 default >> True
X.head()
```

```python
# data split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=1,stratify=y)
```
